# Linear Regression - Canada GDP

### Preprocessing

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='darkgrid')
```

```python
df = pd.read_csv('https://github.com/codebasics/py/raw/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.plot.scatter('year', 'per capita income (US$)', color='red')
plt.show()
```

```python
for x in df.columns:
    df[x].plot.density()
    plt.show()
```

```python
X_train = df.drop('per capita income (US$)', axis=1)
y_train = df['per capita income (US$)']
```

### Modelling

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_, model.intercept_
```

```python
df.plot.scatter('year', 'per capita income (US$)', color='blue')
plt.plot(
    np.array(df['year']),
    model.coef_[0] * np.array(df['year']) + model.intercept_,
    color='red'
)
plt.show()
```

### Evaluation

```python
model.predict(pd.DataFrame({'year':[2020]}))[0]
```
